```
#print(data)
E = data['E'].astype('float64')
x = (data['D']) 
y = E


fig= plt.figure(figsize=(15, 15))
fig = plt.scatter(x, y, s = 10, c = 'red')


z = np.polyfit(x, y, 2)

fit = np.poly1d(z)
print(fit)
yhat = fit(x)       
# or [p(z) for z in x]
ybar = np.sum(y)/len(y)          # or sum(y)/len(y)
ssreg = np.sum((yhat-ybar)**2) 
# or sum([ (yihat - ybar)**2 for yihat in yhat])
sstot = np.sum((y - ybar)**2) 
# or sum([ (yi - ybar)**2 for yi in y])
ssreg / sstot
print(ssreg/sstot)

plt.plot(x, yhat, c = 'green')
#plt.xlim(right = 1.5)
plt.ylabel('Energy')
plt.xlabel('Density')
```
